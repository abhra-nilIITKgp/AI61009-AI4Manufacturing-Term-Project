```python
logs = pd.read_csv('Fetch_Reacher_Training_Logs.txt', delim_whitespace=True, header=None)
```

```python
logs.head()
```

```python
logs.to_csv('Fetch_Reacher_Training_Logs.csv')
```

```python
logs.plot(x="episode_no", y=["average_reward"])
plt.show()
```

```python
logs[ 'median_reward' ] = logs.average_reward.rolling(100).mean()
  
```

```python
logs[ 'max_reward_rolling_avg' ] = logs.max_reward.rolling(100).mean()
```
